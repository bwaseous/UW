## Simple Euler method demo

```python
#euler method simulator

deltat=.001 #timestep
Tmax=10

R=1
C=1

tlist=np.linspace(0,Tmax,int(Tmax/deltat +1))
Vlist=np.zeros(len(tlist))

# initialize
V0=1
Vlist[0]=V0

#perform Euler approximation
for i in range(len(tlist)-1):
    t=tlist[i]
    Vlist[i+1]=Vlist[i] + (Vlist[i])*deltat
    #Vlist[i+1]=Vlist[i] + (np.sin(t)/1 + np.cos(t+3) + (-Vlist[i]/(1*1))*deltat);
    #Vlist[i+1]=Vlist[i] + (-Vlist[i]/(R*C) + np.sin(t)/C )*deltat;


plt.plot(tlist,Vlist,'.', label = "Euler Approx", linewidth=6)
plt.xlabel('t')
plt.ylabel('V(t)')
plt.plot( np.linspace(0,Tmax,1000), np.exp( np.linspace(0,Tmax,1000) ), label = "Exact Solution", linewidth = 2)
plt.legend()
plt.title("Simple Euler's Approximation to dv/dt = v")
plt.show()

```
