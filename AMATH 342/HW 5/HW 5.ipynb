```python
deltat=1 #timestep
Tmax=40

tlist=np.linspace(0,Tmax,int(Tmax/deltat +1))
Vlist1=np.zeros(len(tlist))
gapplist1=np.zeros(len(tlist))

threshold = 10
E=11
R=1
C=10

gbar = 1/5

gapplist1[3] = gbar
gapplist1[4] = gbar

for i in range(len(tlist)-1):
    t=tlist[i]
    Vlist1[i+1]=Vlist1[i] + (-Vlist1[i]/(R*C) + gapplist1[i]*(E-Vlist1[i]/C ))*deltat
    
f_1 = max(Vlist1)/threshold
plt.plot(tlist, Vlist1, '')
plt.title('Conductane Impulse with gbar = {} and f = {}'.format(gbar, f_1))
plt.xlabel('Time (ms)')
plt.ylabel('Membrane Potential (mV)')
plt.show()
```

The nonlinear relationship between f and N is better seen when the conductance impulse is longer, $\Delta = 4$ms wide, in fact, as the more impulses correlates to a nonlinear reduction if the change in f, i.e., as more conductance impulses arrive, the rate at which the maximal voltages are increasing is decreasing. At lower values of f, however, the rate of change is linear which is why using $\frac{1}{f}$ is a good approximation for the number of impulses needed to overcome the threshold.

This relationship between $f$ and $N$ can be proven mathematically using the solution to the RC Circuit Model with Conductance Based Inputs given V(0) = 0.
By the Conductance RC Circuit Model:
$$V(t) = E_{syn}\int^t_0 dt' \ g(t') \times \text{exp}(-\int^t_{t'} \frac{1}{\tau} + g(s) \ ds) $$
To calculate the fractional threshold at N impulses is as follows:
$$f = \frac{V(t)}{10} = \frac{E_{syn}\int^t_0 dt' \ N g(t') \times \text{exp}(-\int^t_{t'} \frac{1}{\tau} + Ng(s) \ ds)}{10} $$
Notice that an N appears in the exponentional. If this N along with the N outside of the integral are drawn out to the front, they become
$$ f = Ne^{-N} \times \int \ldots$$
Thus, the relationship between N and f is a decaying nonlinear relationship. This makes intuitive sense as a linear relationship or even growing nonlinear relationship would mean that the cell, when inundated with a large volume of stimulus, would experience extreme voltage levels which would undoubtedly hurt the cell. Also, conductances are the inverse of resistance, so as more conductances are summed together, the resulting values decreases.

```python
def beta_n(v):
    r =.125*np.exp(-(v+65)/80);
    return r

def beta_m(v):
    r =  4*np.exp(-(v+65)/18);
    return r

def beta_h(v):
    r =  1/(1+np.exp(-(v+35)/10));
    return r

def alpha_n(v):
    r =  .01*(v+55)/(1-np.exp(-(v+55)/10));
    return r

def alpha_m(v):
    r =  .1*(v+40)/(1-np.exp(-(v+40)/10));
    return r

def alpha_h(v):
    r =  .07*np.exp(-(v+65)/20);
    return r
```

```python
vna=50; 
vk=-77
vl=-54.4
gna=120
gk=36
gl=.3
c=1
v_init=-65
m_init=.052
h_init=.596
n_init=.317
```

```python
spikes = np.zeros(50)

for jj in range(len(spikes)):
    
    nsteps=25000
    dt=0.01
    
    v=np.zeros([nsteps])
    m=np.zeros([nsteps])
    n=np.zeros([nsteps])
    h=np.zeros([nsteps])
    t=np.zeros([nsteps])

    m[0]=m_init
    n[0]=n_init
    h[0]=h_init
    v[0]=v_init
    t[0]=0.0
    
    I=4*jj
    spike = 0
    
    iter_arr=np.arange(nsteps-1)
    for step in iter_arr:
        v[step+1]=v[step]+((I - gna*h[step]*(v[step]-vna)*m[step]**3 -gk*(v[step]-vk)*n[step]**4-gl*(v[step]-vl))/c)*dt;
        m[step+1]=m[step]+ (alpha_m(v[step])*(1-m[step])-beta_m(v[step])*m[step])*dt;
        h[step+1]=h[step]+ (alpha_h(v[step])*(1-h[step])-beta_h(v[step])*h[step])*dt;
        n[step+1]=n[step]+ (alpha_n(v[step])*(1-n[step])-beta_n(v[step])*n[step])*dt;
        t[step+1]=t[step]+dt;
        if ((step> 1) & (v[step+1]<v[step]) & (v[step] > v[step-1]) & (v[step] > -25)):
            spike = spike + 1
    spikes[jj] = spike

spikerate = spikes/250
```

```python
current = 4*np.arange(0,50,1)
plt.plot(current, spikerate,'k-')
plt.xlabel('$\overline{I} \ (\mu A)$')
plt.ylabel('Firing Rate (Hz)')
plt.title('Firing Rate of Cell Modeled by HH with Varied Current')
plt.show()
```

As the current gradually increases, a threshold is passed for generating an action potential. After this threshold has been passed, the firing rate will dramatically increase as more stimulus pushes the cell to send a greater response - more action potentials. This causes the firing rate to ramp up, but after a point, dramatically fall as the cell can no longer register higher currents.

This cliff in firing rate could be due to the cell malfunctioning because of the strong current or the machinery of the cell cannot physically emit more action potentials, e.g., it has exhausted its supply of ions to transport across its cell membrane.

The firing rate is calcualted by checking if the voltage before is greater than the current voltage value, and the subsuquent voltage is less than the current voltage.

Based upon the graphs above that change $\epsilon$ and $\omega$, respectively, while keeping the other constant reveal that increasing $\epsilon$ increases the maximal Firing Rate but also increases the threshold at which firing occurs. This squishes the firing rate curve into a shorter range of registered activation in a region of higher current.
