```python
n_batch = 1
n_time = 500
n_bits = 1

inputs = np.zeros([n_batch, n_time, n_bits])
outputs = np.zeros([n_batch, n_time, n_bits])

freq_range = 50
time_length = n_time

for trail_index in np.arange(n_batch):
    freq = freq_range #np.random.randint(4, freq_range + 4)
    const_signal = np.repeat(freq / freq_range + 0.25, time_length)
    const_signal = np.expand_dims(const_signal, axis=1)
    inputs[trail_index,:,0] = const_signal[:,0]
    t = np.arange(0, time_length*0.005, 0.005)
    target = np.sin(freq * t)
    target = np.expand_dims(target, axis=1)
    outputs[trail_index,:,0] = target[:,0]
        
plt.plot(target)
plt.plot(const_signal)
plt.show()
        

for i in range(n_batch):
    plt.plot(inputs[i,:,0], 'b-')
    plt.plot(outputs[i,:,0], 'r-')
```
